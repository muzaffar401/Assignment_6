## ðŸ”¥ Python Code Example Step-by-Step:

### Step 1: Ek simple `Class` banao

```python
class Car:
    pass
```

âž¡ï¸ Abhi humne sirf ek **design** banaya hai. Isme kuch nahi hai.

---

### Step 2: Class ke andar **Attributes** aur **Methods** banao

```python
class Car:
    # Constructor banate hain (automatically chalne wala function)
    def __init__(self, color, speed):
        self.color = color      # Attribute 1
        self.speed = speed      # Attribute 2

    # Method: gadi start karne ka kaam
    def start(self):
        print(f"Car is starting... Color: {self.color}, Speed: {self.speed} km/h")

    # Method: gadi stop karne ka kaam
    def stop(self):
        print("Car is stopping...")
```

**Samajh:**  
- `__init__` function har nayi car banate waqt chalta hai.
- `self` ka matlab hai: "yeh car ka khud ka data".

---

### Step 3: Ek `Object` banao (Asli car)

```python
my_car = Car("Red", 200)
```

âž¡ï¸ Ab `my_car` ek asal car hai jiska color Red aur speed 200 hai.

---

### Step 4: Methods call karo

```python
my_car.start()
my_car.stop()
```

**Output:**
```
Car is starting... Color: Red, Speed: 200 km/h
Car is stopping...
```

## ðŸ”¥ Pehle samajho: Constructor kya hota hai?

- Constructor ek **special function** hota hai class ke andar.
- Iska naam Python mein **hamesha `__init__`** hota hai.
- Jab hum **object banate hain**, to **yeh function automatic chal jata hai**.

Jaise:

```python
class Car:
    def __init__(self, color, speed):
        self.color = color
        self.speed = speed
```

---

## ðŸ”¥ Aapka Sawal:  
**"Direct property kyun nahi likhte? Constructor kyun?"**

Haan! Theek socha aapne, hum direct bhi likh sakte hain.  
Jaise:

```python
class Car:
    color = "Red"
    speed = 200
```

Lekin... **is tarah problem hoti hai!**  
Wo kya?

---

## ðŸ”¥ Wajah (Reason) â€” Why Constructor Zaroori Hai:

| Point | Direct property | Constructor |
|:--|:--|:--|
| 1 | **Sab objects me same data hoga** | **Har object apna alag data le sakta hai** |
| 2 | **Flexibility nahi** | **Zyada flexibility milti hai** |
| 3 | **Hard coded hota hai** | **User se ya input se data le sakte hain** |
| 4 | **Limited use** | **Real-life cheeze dynamic hoti hain** |

---

### ðŸ›‘ Problem with Direct Properties Example:

```python
class Car:
    color = "Red"
    speed = 200

car1 = Car()
car2 = Car()

print(car1.color)  # Red
print(car2.color)  # Red
```

âœ… Kaam to karega, **lekin dono cars same honge**!  
**Har car ka color aur speed same honge**, jo real life mein galat hai.

---

### âœ… Constructor se Dynamic Objects bana sakte hain:

```python
class Car:
    def __init__(self, color, speed):
        self.color = color
        self.speed = speed

car1 = Car("Red", 200)
car2 = Car("Blue", 180)

print(car1.color)  # Red
print(car2.color)  # Blue
```

**Dekha?**
- `car1` red color aur 200 speed wali hai
- `car2` blue color aur 180 speed wali hai

**âž¡ï¸ Har object alag alag properties ke sath bana!**

## ðŸ”¹ 1. Association Kya Hai?

Association ka matlab hota hai:

> â€œDo classes ka aapas me **sirf relation** hai, **ownership nahi**.â€

Yani:

* Ek class dusri class ke object ko **use** to karti hai...
* Lekin **uski malik nahi hoti**.
* Dono alag alag bhi zinda reh sakte hain.

---

## ðŸ”¸ Real-Life Example:

### **Teacher aur Student**

* Teacher apna kaam kar sakta hai.
* Student bhi apna kaam kar sakta hai.
* Lekin kabhi kabhi student ka teacher hota hai = **association**.

Dono **independent** hain. Teacher chala bhi jaye to student phir bhi zinda hai.

---

## ðŸ”¸ Python Example:

```python
class Teacher:
    def __init__(self, name):
        self.name = name

class Student:
    def __init__(self, name):
        self.name = name
        self.teacher = None  # initially no teacher

    def set_teacher(self, teacher):
        self.teacher = teacher

# ðŸ‘‡ Objects bana rahe hain
t1 = Teacher("Sir Ali")
s1 = Student("Ahmed")

# ðŸ‘‡ Ahmed ka teacher set kar rahe hain
s1.set_teacher(t1)

# ðŸ‘‡ Output dekhte hain
print(f"{s1.name} ka teacher hai {s1.teacher.name}")
```

---

## ðŸ”¹ Output:

```
Ahmed ka teacher hai Sir Ali
```

---

## ðŸ§  Soch Samjho:

* `Teacher` class me `Student` ka koi object nahi.
* `Student` class me `Teacher` ka sirf **reference** (link) hai.
* Dono class apni jagah **independent** hain.
* Sirf **relation banaya gaya**, koi "ownership" nahi hai.

---

## ðŸ”¶ Aggregation Story Style â€“ â€œSchool & Studentsâ€

## ðŸ”· Coding Example (super asaan):

```python
class Student:
    def __init__(self, name):
        self.name = name

class School:
    def __init__(self, school_name):
        self.school_name = school_name
        self.students = []  # yahan students rakhe jaate hain

    def add_student(self, student):
        self.students.append(student)

# ðŸŽ“ Pehle se student exist karte hain
s1 = Student("Ahmed")
s2 = Student("Zara")

# School banaya
my_school = School("Happy Public School")

# Students ko school me add kiya
my_school.add_student(s1)
my_school.add_student(s2)

# School aur uske students ka naam print karo
print("School:", my_school.school_name)
for student in my_school.students:
    print("Student:", student.name)
```

## ðŸ”´ Composition Kya Hai?

> **"Strongest relationship"** hota hai â€” ek class doosri class ka **part** hoti hai aur **us ke bina zinda nahi reh sakti.**

### ðŸ“Œ Short Formula:

```
Agar parent mar gaya to child bhi mar gaya = Composition
```

---

## ðŸ”¸ Real-Life Example:

### ðŸ§â€â™‚ï¸ Human aur Heart

* Insaan ke paas **heart (dil)** hota hai.
* Agar insaan mar jaye, to uska heart bhi zinda nahi rehta.
* Heart **insaan ka part hai** = **composition**

---

## ðŸ”· Python Example: Human â€“ Heart

```python
class Heart:
    def __init__(self):
        self.status = "Beating"

    def beat(self):
        print("Dhak Dhak...")

class Human:
    def __init__(self, name):
        self.name = name
        self.heart = Heart()  # ðŸ‘ˆ Heart object yahin ban raha hai (composition)

    def show_heart(self):
        print(f"{self.name} ka heart:", self.heart.status)
        self.heart.beat()

# ðŸ‘¨ Human object bana
h1 = Human("Ali")
h1.show_heart()
```

---

## ðŸ”¹ Output:

```
Ali ka heart: Beating
Dhak Dhak...
```

---

### â— Important Point:

* `Heart` object ko `Human` class **khud create** kar rahi hai.
* `Heart` object **sirf Human ke through zinda** hai.
* Agar `h1` delete ho gaya:

```python
del h1
```

To uska `heart` bhi gaya â€” kyunki woh uska **part** tha.
Isko bolte hain **Composition**.

## ðŸ”· 1. Inheritance Kya Hai? (Asaan Lafzon Me)

> Jab ek class (child) doosri class (parent) ke **functions** aur **variables** ko **reuse** karti hai, to usay **inheritance** kehte hain.

ðŸ§  Isse:

* Code reuse hota hai
* Logical relation dikhaya ja sakta hai (jaise: **Car is a Vehicle**)

---

## ðŸ”· 2. Real Life Example

### ðŸ‘¨â€ðŸ« Teacher â†’ ðŸ‘¨â€ðŸŽ“ Student

* **Teacher** ke paas knowledge aur bolne ki ability hai.
* **Student** unhi skills ko seekhta hai aur use karta hai, aur apne khud ke skills bhi develop karta hai jaise exam dena.

> Iska matlab: Student ne Teacher se kuch cheezen **inherit** karli, aur kuch apni bhi bana li.

---

## ðŸ”· 3. Python Basic Example

```python
# Parent class
class Teacher:
    def teach(self):
        print("Teaching students")

# Child class
class Student(Teacher):
    def give_exam(self):
        print("Giving exam")
```

### âœ… Use:

```python
s = Student()
s.teach()        # Teacher ka function (inherited)
s.give_exam()    # Student ka apna function
```

---

## ðŸ”· 4. Output:

```
Teaching students
Giving exam
```

## ðŸ”· 1. **Single Inheritance**

> Ek class A hai (Parent), ek class B hai (Child)

```
A â†’ B
```

### ðŸ§  Example:

```python
class A:
    def show_a(self):
        print("This is class A")

class B(A):  # B inherits A
    def show_b(self):
        print("This is class B")
```

### âœ… Use:

```python
obj = B()
obj.show_a()  # Parent ka function
obj.show_b()  # Child ka function
```

### ðŸ“¤ Output:

```
This is class A  
This is class B
```

---

## ðŸ”· 2. **Multilevel Inheritance**

> A se B inherit karta hai, phir B se C.

```
A â†’ B â†’ C
```

### ðŸ§  Example:

```python
class A:
    def show_a(self):
        print("Class A")

class B(A):
    def show_b(self):
        print("Class B")

class C(B):
    def show_c(self):
        print("Class C")
```

### âœ… Use:

```python
obj = C()
obj.show_a()
obj.show_b()
obj.show_c()
```

### ðŸ“¤ Output:

```
Class A  
Class B  
Class C
```

---

## ðŸ”· 3. **Multiple Inheritance**

> A aur B dono parent hain, C child hai

```
A + B â†’ C
```

### ðŸ§  Example:

```python
class A:
    def show_a(self):
        print("A's method")

class B:
    def show_b(self):
        print("B's method")

class C(A, B):  # Inherits both A and B
    def show_c(self):
        print("C's method")
```

### âœ… Use:

```python
obj = C()
obj.show_a()
obj.show_b()
obj.show_c()
```

### ðŸ“¤ Output:

```
A's method  
B's method  
C's method
```

---

## ðŸ”· 4. **Hierarchical Inheritance**

> A ek parent hai, B aur C dono us se inherit karte hain.

```
    A
   / \
  B   C
```

### ðŸ§  Example:

```python
class A:
    def show_a(self):
        print("I am class A")

class B(A):
    def show_b(self):
        print("I am class B")

class C(A):
    def show_c(self):
        print("I am class C")
```

### âœ… Use:

```python
b = B()
b.show_a()
b.show_b()

c = C()
c.show_a()
c.show_c()
```

### ðŸ“¤ Output:

```
I am class A  
I am class B  
I am class A  
I am class C
```

---

## ðŸ”· 5. **Hybrid Inheritance**

> Mix of multiple + multilevel

```
    A
   / \
  B   C
   \ /
    D
```

### ðŸ§  Example:

```python
class A:
    def show_a(self):
        print("A method")

class B(A):
    def show_b(self):
        print("B method")

class C(A):
    def show_c(self):
        print("C method")

class D(B, C):  # Hybrid: inherits B and C (both inherited A)
    def show_d(self):
        print("D method")
```

### âœ… Use:

```python
obj = D()
obj.show_a()  # From A
obj.show_b()  # From B
obj.show_c()  # From C
obj.show_d()  # From D
```

### ðŸ“¤ Output:

```
A method  
B method  
C method  
D method
```
